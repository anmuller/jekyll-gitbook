```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 400)

# 2. Calculer les valeurs correspondantes de y
y = x ** 2

# 3. Tracer le graphique
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2", color="blue")
plt.title("Graphique de la fonction $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
